# Series in pandas

List to Series

```python
data = [1,2,3,4,5,6,7,8]

S1 = pd.Series(data)                # default integer indexing
print(S1)

indices = ['a','b','c','d','e','f','g','h']

S2 = pd.Series(data,indices)        # user defined indexing
S2
```

Access values

```python
print(S2['g'])
print(S1[4])
```

Dictionary to Series

```python
d1 = {'oranges':5,'mango':3,'banana':7,'apple':2}
S3 = pd.Series(d1)
print(S3,type(S3))
```

# DataFrames in pandas

```python
d2 = {'Name':['neeraj','harsh','aman','krish'],'Age':[20,23,22,19]}
df1 = pd.DataFrame(d2)
df1
df2 = pd.DataFrame(d2,index=['rank 3','rank 1','rank 2','rank 4'])
df2
```

Naming for attributes later

```python
d3 = [['neeraj',20],['harsh',23],['aman',22],['krish',19]]
df3 = pd.DataFrame(d3,columns=['Name','Age'])
df3
```

Create own random dataset??

```python
import pandas as pd
import random as r
import string as s

df = pd.DataFrame({'ID':     r.sample(range(1,21),20),
                   'NAME':   ["".join(r.sample(s.ascii_letters,5)) for _ in range(20)],
                   'AGE':    [r.randint(15,80) for _ in range(20)],
                   'Branch': [r.choice(['COPC','COE','ENC','ECE']) for _ in range(20)],
                   'CGPA':   [round(r.uniform(5,10),2) for _ in range(20)],
                   'gender': ["male" if r.random()>0.5 else "female" for _ in range(20)]
                   })
df
```

Set date as index

```python
import numpy as np
dates = pd.date_range(start='2025-07-24',end='2025-08-20',periods=6)
dates
dff = pd.DataFrame(np.random.randn(6,4),index=dates.date,columns=list('ABCD'))
dff
```

Sort Dataset

```python
dff.sort_index(axis=1,ascending=False)
# axis = 1 : pick up the column indices(as well as values) and sort
# axis = 0 : pick up the row indices(as well as values) and sort
```

Concatenate DataFrames

```python
df1 = pd.DataFrame(np.random.randn(5,4))
print(df1)
df2 = pd.DataFrame(np.random.randn(6,3))
print(df2)
df3 = pd.concat([df1,df2],axis=0,ignore_index=True)
# df3.reset_index(drop=True,inplace=True)
# No need of reset if ignore_index is True
print(df3)
```
